### Собираем с первых пяти страниц

```python
from time import sleep
```

```python
data = []

for page in range(1,6):
    print(f'Page {page}')
    url = f'https://shop.mts.by/phones/?page={page}'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    telephones = soup.findAll('div', class_='products__unit')
    for telephone in telephones:
        model = telephone.find('div', class_='products__unit__title').find('a').text
        price = telephone.find('span', class_='products__unit__price__number products__unit__price__number--full').text
        price_monthly = telephone.find('span', class_='products__unit__price__number products__unit__price__number--monthly').text
        link = 'https://shop.mts.by' + telephone.find('div', class_='products__unit__title').find('a').get('href')
        #print(model, price, price_monthly, link)

        data.append([model, price, price_monthly, link])
    sleep(5)
```
